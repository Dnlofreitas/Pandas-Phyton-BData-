```python
import pandas as pd
```

```python
TabelaJAS = '/GAM FIAPE - 23.09.xlsx'
df = pd.read_excel(TabelaJAS)
```

```python
df['Container Numbers'] = df['Container Numbers'].str.split(',')
```

```python
df_expanded = df.explode('Container Numbers')
```

```python
df_expanded['Actual Pickup'] = pd.to_datetime(df_expanded['Actual Pickup'], errors='coerce')
```

```python
df_expanded['Pick Up Week'] = df_expanded['Actual Pickup'].dt.isocalendar().week
```

```python
df_cleaned = df_expanded[['Consignee Code','Consignor Name','Consol Transport Mode','Container Numbers','Carrier Name','Pick Up Week','Actual Pickup','First Air/Sea Port','First Leg Arrival Port','Second Air/Sea Port','Second Air/Sea DischargePort','Vessel','ETD First Load','Leg 1 ATA','Vessel (Last Leg)','Leg 2 ATD','Second Air/Sea ATA']]
```

```python
df_cleaned = df_cleaned.drop_duplicates(subset=['Pick Up Week', 'First Air/Sea Port', 'Second Air/Sea Port',
                                                'Second Air/Sea DischargePort', 'Vessel', 'Vessel (Last Leg)'],
                                        keep='first')
```

```python
df['Container Numbers'] = df['Container Numbers'].apply(lambda x: [item.strip() for item in x] if isinstance(x, list) else x)
```

```python
#erro#
df['Consol Transport Mode'] = df['Consol Transport Mode'].str.replace('AIR', '', regex=False)
```

```python
#erro#
df_expanded['Actual Pickup'] = df_expanded['Actual Pickup'].dt.normalize()
```

```python
df_cleaned.to_excel('TabelaJASRevised.xlsx', index=False)
```
